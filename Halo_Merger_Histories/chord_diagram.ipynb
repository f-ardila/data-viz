```python
def get_closest_halo_from_masses(proposed_masses, a):
    
    log_masses = np.log10(a['mvir'])
    
    indices = [np.argmin(np.abs(m-log_masses)) for m in proposed_masses]
    
    halos = [a[i] for i in indices]
    
#     print( [np.log10(h['mvir']) for h in halos])
    return halos
```

```python
class InvalidInputError(Exception):
    pass
def get_idx_interv(d, D):
    k=0
    while(d>D[k]): 
        k+=1
    return  k-1
def deCasteljau(b,t): 
    N=len(b) 
    if(N<2):
        raise InvalidInputError("The  control polygon must have at least two points")
    a=np.copy(b) #shallow copy of the list of control points 
    for r in range(1,N): 
        a[:N-r,:]=(1-t)*a[:N-r,:]+t*a[1:N-r+1,:]                             
    return a[0,:]

def BezierCv(b, nr=5):
    t=np.linspace(0, 1, nr)
    return np.array([deCasteljau(b, t[k]) for k in range(nr)]) 
```

```python
def plot_nodes_in_circle(halo_masses, log=False):
    
    n=len(halo_masses)
#     halo_masses = np.sort(halo_masses)
    
    if log:
        halo_masses=np.log10(halo_masses)
        scale_factor = 10
        labels=np.round(halo_masses,2)
    else:
        scale_factor=20
        labels=np.round(np.log10(halo_masses),2)
    
    
    
    # create a color palette, mapped to these values
    mini=min(halo_masses)
    maxi=max(halo_masses)
    norm_color = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
    node_colors = [norm_color(value) for value in halo_masses]

    #scale sizes
#     scaled_sizes =( halo_masses/max(halo_masses))*scale_factor
    norm_size = matplotlib.colors.Normalize(vmin=min(halo_masses)-1, vmax=max(halo_masses))
    scaled_sizes = [norm_size(value)*scale_factor for value in halo_masses]

    coords = unit_circle_xy_coordinates_for_n_points(n)
    Xn=[c[0] for c in coords]
    Yn=[c[1] for c in coords]

    # plot nodes
    trace2=go.Scatter(x=Xn,
               y=Yn,
               mode='markers',
               name='',
               marker=dict(symbol='circle',
                             size=scaled_sizes,
                             sizemin=2,
#                              sizeref=100,
#                              sizemode='area',
                             color=node_colors,
                             colorscale='Rainbow',
#                              colorbar=dict(thickness=20, tickmode='array', tickvals = np.linspace(0,1,6),
#                                            ticktext=np.round(np.log10(norm_color.inverse(np.linspace(0,1,6)).data),2))
#                              line=dict(color=line_color, width=0.5)
                             ),
               text=labels,
               hoverinfo='text',
               )

    data=[trace2]
    
    return data, node_colors
```

```python
def make_chord_diagram_given_halo(halo, max_z, sort_order, log):
    
    #use halos within redshift 0.1
    halo_mask = np.array(halo['tree', 'redshift']<max_z)

    
    #get masses
    halo_masses = np.array(halo['tree', 'mvir'])[halo_mask]
    
    #get redshifts
    halo_redshifts = np.array(halo['tree', 'redshift'])[halo_mask]
    
        
    if sort_order == 'mass':
        sorted_indices = np.argsort(halo_masses)
        
    elif sort_order == 'z':
        sorted_indices = np.argsort(-halo_redshifts) #reverse order
    
    halos = halo['tree'][halo_mask]
    print(len(halos), ' halos')
    #plot nodes
    nodes, node_colors = plot_nodes_in_circle(halo_masses[sorted_indices], log=log)
    print("NODES!")
    
    #get edges
    edges = get_edges_for_halos(halos[sorted_indices], max_z)
    print("EDGES!")
#     print(sorted(edges, key=lambda tup: tup[0]))
    
    edges_layout=unit_circle_xy_coordinates_for_n_points(len(halo_masses))
    
    #plot edges
    lines = plot_edges(edges, edges_layout, node_colors)
    print("LINES!")
    
    #layout
    layout = make_layout(halo, max_z)
    
    return nodes, lines, layout
    
#     fig=go.Figure(data=nodes+lines, layout=layout)
#     py.iplot(fig, filename='Halos')
```
