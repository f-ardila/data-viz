```python
def dataframe_values(df):
    
    dictionary = dict()

    for question in df.keys():
        dictionary[question] = dict()
        
#         possible_answers = set(df[question])
        
        possible_answers = ['Emerging', 'Developing', 'Transforming', "I don't know"]
        
        for answer in possible_answers:
            value = np.sum(df[question].str.startswith(answer))
            dictionary[question][answer] = value
        
    
    return dictionary
```

```python
grad_values = dataframe_values(grads)
faculty_values = dataframe_values(faculty)
total_values = dataframe_values(df)
```

```python
def func(pct, allvals):
    if pct == 0:
        return
    allvals = np.array(list(allvals))
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} votes)".format(pct, absolute)
```

```python
size=0.3
cmap = plt.get_cmap("tab20")
inner_colors = cmap(np.arange(4)*2)
outer_colors = cmap((np.arange(4)*2)+1)


fig, axes = plt.subplots(1,3, figsize=[15,5])

axes[0].pie(total1.values(), autopct=lambda pct: func(pct, total1.values()),
            radius=1, colors=inner_colors,
            wedgeprops=dict( edgecolor='w'), textprops=dict(color="w"))
axes[0].set(aspect="equal", title='All')


axes[1].pie(grads1.values(), autopct=lambda pct: func(pct, grads1.values()),
            radius=1-size, colors=inner_colors,
            wedgeprops=dict(edgecolor='w'), textprops=dict(color="w"))
axes[1].pie(total1.values(), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))
axes[1].set(aspect="equal", title='Grads')

axes[2].pie(faculty1.values(), autopct=lambda pct: func(pct, faculty1.values()),
            radius=1-size, colors=inner_colors,
            wedgeprops=dict(edgecolor='w'), textprops=dict(color="w"))
axes[2].pie(total1.values(), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))
axes[2].set(aspect="equal", title='Faculty')


axes[0].legend(loc='upper left', labels = ['Emerging', 'Developing', 'Transforming', "I don't know"])
fig.tight_layout()

plt.show()
```

```python
fig, ax = plt.subplots()

size = 0.3
vals = np.array([[60., 32.], [37., 40.], [29., 10.]])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9, 10]))

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title='Pie plot with `ax.pie`')
plt.show()
```
